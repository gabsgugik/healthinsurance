```python
class Integrantes:
    def __init__( self, nome, peso, altura):
        self.nome = nome
        self.peso = peso
        self.altura = altura
        self.salario = 10

    def get_salario( self ):
        return self.salario
```

```python
class Jogador( Integrantes ):
    def __init__( self, nome, peso, altura, gol ):
        self.gol = gol
        super().__init__(nome, peso, altura)
        
    def get_salario( self ):
        total = super().get_salario() * self.gol
        return total
        
```

```python
meigarom = Jogador('Meigarom', 88, 1.8, 5)
meigarom.get_salario()
```
